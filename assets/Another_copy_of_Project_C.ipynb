```r

# Function to calculate p-values and significance of correlations
assess_correlation_significance <- function(correlation_matrix) {
  p_values <- matrix(NA, nrow = ncol(correlation_matrix), ncol = ncol(correlation_matrix))
  for (i in 1:ncol(correlation_matrix)) {
    for (j in 1:ncol(correlation_matrix)) {
      if (i != j) {
        cor_test <- cor.test(correlation_matrix[, i], correlation_matrix[, j])
        p_values[i, j] <- cor_test$p.value
      }
    }
  }
  return(p_values)
}

# Calculate p-values and significance of correlations
p_values <- assess_correlation_significance(correlation_matrix)

# Print p-values
print(p_values)
```

```r
# Load necessary libraries
library(ggplot2)


# Assuming your data frame is named data
# Check the structure of your data

# List of continuous variables to check
continuous_vars<-c("TST","TSTR", "DEPRESSION", "ANXIETY", "STRESS")
# Function to plot histograms and Q-Q plots
plot_normality <- function(mental_data, var) {
  # Histogram
  ggplot(data, aes_string(var)) +
    geom_histogram(aes(y = ..density..), bins = 30, fill = "blue", alpha = 0.5) +
    geom_density(color = "red") +
    ggtitle(paste("Histogram of", var)) +
    theme_minimal()

  # Q-Q plot
  ggplot(data, aes_string(sample = var)) +
    stat_qq() +
    stat_qq_line() +
    ggtitle(paste("Q-Q Plot of", var)) +
    theme_minimal()
}

# Apply the function to each continuous variable
for (var in continuous_vars) {
  print(plot_normality(mental_data, var))
}

# Function to perform Shapiro-Wilk test
shapiro_test <- function(mental_data, var) {
  shapiro.test(mental_data[[var]])
}

# Apply the Shapiro-Wilk test to each continuous variable and print results
for (var in continuous_vars) {
  test_result <- shapiro_test(mental_data, var)
  print(paste("Shapiro-Wilk test for", var, ":"))
  print(test_result)
}

# Function to perform Kolmogorov-Smirnov test
ks_test <- function(mental_data, var) {
  # Test against normal distribution with mean and sd of the data
  ks.test(mental_data[[var]], "pnorm", mean = mean(mental_data[[var]], na.rm = TRUE), sd = sd(mental_data[[var]], na.rm = TRUE))
}

# Apply the KS test to each continuous variable and print results
for (var in continuous_vars) {
  test_result <- ks_test(mental_data, var)
  print(paste("Kolmogorov-Smirnov test for", var, ":"))
  print(test_result)
}
```

t test cant be used as the continous variables are not normally distributed

for data not normally distributed non parametric test *Mann Wittney U test* is applied

```r
# Load necessary library
library(stats)

# Define a function to perform Mann-Whitney U test
perform_mann_whitney_test <- function(data, variable, group) {
  result <- wilcox.test(data[[variable]] ~ data[[group]], data = data)
  return(result)
}

# List of continuous variables
continuous_vars <- c("TST", "TSTR", "DEPRESSION", "ANXIETY", "STRESS")

# Perform Mann-Whitney U test for each variable
for (var in continuous_vars) {
  test_result <- perform_mann_whitney_test(data, var, "Gender")
  cat("\nMann-Whitney U test for", var, ":\n")
  print(test_result)
}
```

```r
# Load your primary data
primary_data <- c(STST)

# Visualize the original data distribution
hist(primary_data, breaks = 30, main = "Original Data Distribution", xlab = "Data", col = "lightblue", border = "black")

# Apply the Box-Cox transformation
bc <- boxcox(primary_data ~ 1, lambda = seq(1, 5, by = 0.1)) # Using a range of lambda values
best_lambda <- bc$x[which.max(bc$y)] # Extracting the best lambda
transformed_data <- (primary_data^best_lambda - 1) / best_lambda

# Visualize the transformed data distribution
hist(transformed_data, breaks = 30, main = "Transformed Data Distribution", xlab = "Transformed Data", col = "lightgreen", border = "black")

```
